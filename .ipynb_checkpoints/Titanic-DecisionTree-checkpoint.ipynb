### EDA

#### Features' Distribution
Histogram for continuous variables and bar chart for categoricals, by 'Survived'.

```python
fig , ax = plt.subplots(4,2,figsize = (10,20))     # set up 2 x 2 frame count plot with figsize 10 x 10

all_list = ['Age','SibSp','Parch','Fare','Pclass','Sex','Cabin','Embarked']
num_list = all_list[:4]
cat_list = all_list[4:]

for column , splots in zip (all_list, ax.flatten()):  

    if column in num_list:
        col_max = df[column].max()
        num_plot = sns.distplot(df[df['Survived']==0][column],ax = splots, kde = False, 
                     bins = list(np.arange(0,col_max,col_max/10)), 
                     hist_kws=dict(edgecolor="black", linewidth=1), label = '0')
        num_plot = sns.distplot(df[df['Survived']==1][column],ax = splots, kde = False, 
                     bins = list(np.arange(0,col_max,col_max/10)), 
                     hist_kws=dict(edgecolor="black", linewidth=1), label = '1');
        num_plot.set_xlim(0,col_max)
        num_plot.legend(title='Survived')
    elif column in cat_list:
        sns.countplot(df[column],hue = df['Survived'],ax = splots, palette = 'BuPu',
                     order=df[column].value_counts().iloc[:10].index);
```

```python
fig, axs = plt.subplots(3, 3, figsize=(14,14))

for column in enumerate(df.columns):
        sns.histplot(data=df, x=column[1], kde=True, ax=axs[(column[0])//3, (column[0])%3])
    
plt.show()
```

##### Age distribution
We look at the age distribution subsets to gain information for treating the missing values issue.

```python
age_female = df[(df['Sex']=='female')&(df['Age'].notnull())]['Age']
age_male = df[(df['Sex']=='male')&(df['Age'].notnull())]['Age']
```

```python
fig, axs = plt.subplots(ncols=2)
fig.set_size_inches(14, 6)

plot1 = sns.distplot(age_female, ax=axs[0])
plot1.set(title='female')

plot2 = sns.distplot(age_male, ax=axs[1])
plot2.set(title='male')
```

##### Outliers
Histogram plot of the age suggests the existence of outliers. However, the data points outside our boxplot bounderies are not distance enough to be justified as outliers.

```python
fig, axs = plt.subplots(ncols=2)
fig.set_size_inches(14, 6)

plot1 = sns.boxplot(age_female, ax=axs[0])
plot1.set(title='female')

plot2 = sns.boxplot(age_male, ax=axs[1])
plot2.set(title='male')
```

##### Missing Values
'Cabin' has too many missing values. So we will remove it for this project. 
Based on the skewed graph of 'Age' we choose median as the imputer. We also use mode for the 2 missing 'Cabin' values.

```python
print('female age, median:',age_female.median())
print('male age, median:',age_male.median())
print('Embarked, mode:', df.mode()['Embarked'][0])
```

```python
df['Embarked'].mode()
```

```python
df_m = df.drop('Cabin',axis=1).copy(deep=True)
df_m.loc[(df['Sex']=='female')&(df['Age'].isnull()),'Age'] = age_female.median()
df_m.loc[(df['Sex']=='male')&(df['Age'].isnull()),'Age'] = age_male.median()
df_m.loc[(df['Embarked'].isnull()), 'Embarked'] = df.mode()['Embarked'][0]
```

```python
df_m.isna().sum()
```

### Data Preparation for Decision Tree

```python
for column in df[['Survived','Pclass','Sex','Cabin','Embarked']].columns:
    df[column]=df[column].astype('category')
```

```python
enc = OrdinalEncoder()
enc.fit(df_m[['Survived', 'Pclass','Sex','Embarked']])
df_m[['Survived', 'Pclass','Sex','Embarked']] = enc.transform(df_m[['Survived', 'Pclass','Sex','Embarked']])
```

### The Tree
Creating a decision tree model using DecisionTreeClassifier from sklearm. All hyperparameters are left as the default.

```python
n_model = DecisionTreeClassifier()
n_y_pred = n_model.fit(df_m.iloc[:,1:],df_m.iloc[:,0]).predict(df_m.iloc[:,1:])
print(classification_report(df_m.iloc[:,0], n_y_pred))
```

```python
fig = plt.figure(figsize=(30,20))
tree.plot_tree(n_model, feature_names=df_m.iloc[:,1:].columns);
```

### Decision Tree Hyperparameter Tuning
Decision Tree's hyperparameters:
<br>https://towardsdatascience.com/how-to-tune-a-decision-tree-f03721801680<br>
https://ken-hoffman.medium.com/decision-tree-hyperparameters-explained-49158ee1268e

```python
t_model = DecisionTreeClassifier()

params={"splitter":["best","random"],
        'max_depth': list(range(3, 10)),
        "min_samples_leaf": list(range(1, 5)),
        'max_features': ["auto","log2","sqrt",None],
        'min_samples_split': [2, 5, 10],
        "max_leaf_nodes":list(range(2, 10))}

trsearch = RandomizedSearchCV(estimator=t_model, param_distributions=params, scoring = "accuracy",
                              n_iter = 200, cv=10)
```

```python
t = time.process_time()  

trsearch.fit(df_m.iloc[:,1:],df_m.iloc[:,0])

print("Operation time (min): " , (time.process_time()-t)/60)
```

```python
trsearch.best_params_
```

```python
bt_y_pred = t_model.fit(df_m.iloc[:,1:],df_m.iloc[:,0]).predict(df_m.iloc[:,1:])
print(classification_report(df_m.iloc[:,0], bt_y_pred))
```

```python
fig = plt.figure(figsize=(30,20))
tree.plot_tree(t_model, feature_names=df_m.iloc[:,1:].columns);
```

### Bagging Classifier
Bootsraps Aggregating, or Bagging
https://machinelearningmastery.com/bagging-ensemble-with-python/<br>
Creating multiple trees on subsets of the data.

```python
dt_model=DecisionTreeClassifier()
bc_model=BaggingClassifier(dt_model)
bc_y_pred=bc_model.fit(df_m.iloc[:,1:],df_m.iloc[:,0]).predict(df_m.iloc[:,1:])
print(classification_report(df_m.iloc[:,0],bc_y_pred))
```

### Random Forest
https://machinelearningmastery.com/random-forest-ensemble-in-python/<br>
Similar to bagging; with the addition of selecting a random subset of features as well.

```python
from sklearn.ensemble import RandomForestClassifier

cv = KFold(n_splits=10, random_state=60, shuffle=True)
rf_model = RandomForestClassifier()
rf_model_fit = rf_model.fit(df_m.iloc[:,1:],df_m.iloc[:,0])
y_pred = rf_model_fit.predict(df_m.iloc[:,1:])
print(classification_report(df_m.iloc[:,0],y_pred))
```

```python
fig = plt.figure(figsize=(150,60))
tree.plot_tree(rf_model_fit.estimators_[0],filled=True, feature_names=df_m.iloc[:,1:].columns);
```

### Random Forest Hyperparameter Tuning
https://medium.com/analytics-vidhya/random-forest-classifier-and-its-hyperparameters-8467bec755f6<br>
https://towardsdatascience.com/cross-validation-and-hyperparameter-tuning-how-to-optimise-your-machine-learning-model-13f005af9d7d

```python
r_model = RandomForestClassifier()

params={'bootstrap': [True, False],
        'max_depth': list(range(3, 10)),
        "min_samples_leaf": list(range(1, 5)),
        'max_features': ["auto","log2","sqrt",None],
        'min_samples_split': [2, 5, 10],
        "max_leaf_nodes":list(range(2, 10)),
        'n_estimators': [80, 100, 120]}

rr_rsearch = RandomizedSearchCV(estimator=r_model, param_distributions=params, scoring = "accuracy", 
                                n_iter = 200, cv=10)
```

```python
t = time.process_time()  

rr_rsearch.fit(df_m.iloc[:,1:],df_m.iloc[:,0])

print("Operation time (min): " , (time.process_time()-t)/60)
```

```python
br_model = rr_rsearch.best_estimator_
br_model
```

```python
rr_rsearch.best_params_
```

```python
r_model_fit = r_model.fit(df_m.iloc[:,1:],df_m.iloc[:,0])
y_pred = r_model_fit.predict(df_m.iloc[:,1:])
print(classification_report(df_m.iloc[:,0],y_pred))
```

```python
fig = plt.figure(figsize=(30,20))
tree.plot_tree(r_model_fit.estimators_[0],filled=True, feature_names=df_m.iloc[:,1:].columns);
```
